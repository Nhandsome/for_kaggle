## 2. Light GBM
  - [참고](https://nurilee.com/2020/04/03/lightgbm-definition-parameter-tuning/)
  - 트리기반 학습모델
    - 수직적으로 확장하는 경향 : Leaf-wise
      - 다른 Tree 모델들은 Level-wise, 수평적 확장을 한다.
      - 가장 로스를 줄일 수 있는 Leaf를 선택하여 확장 : 최적화 위주
    - Light : 큰 데이터셋에서도 빠른 학습을 진행한다.
      - 최적화 위주인 만큼 작은 데이터셋에서는 과적합하는 취약점을 가진다.
  - 파라미터
    - max_depth : 깊이를 조절, **깊으면 과적합되기 쉽다.**
    - min_data_in_leaf : Leaf가 다루는 레코드의 수, **과적합을 해결**
    - feature_fraction : Random Forest에서 사용, **얼마만큼의 Random 파라미터로 트리를 생성하는가**
    - bagging_fraction : **데이터를 여러개의 Bag에 담에서 나눠서 학습**
    - min_gain_to_split : **분기하기 위한 gain설정**
  - 핵심 파라미터
    - Task : train / predict
    - application : regression / binary / multiclass
    - boosting : gdbt / rf / dart / goss
    - num_boost_round : boosting iteration, over 100
    - learning_rate
    - num_leaves : 전체 Leaf의 수, 디폴트 31
    - device : cpu / gpu
    - metric : mae / mse / binary_logloss / multi_logloss
  - IO 파라미터
    - max_bin : feature를 최대 몇개의 bin으로 나누는가
    - categorical_feature : 카테고리칼 
    - ignore_column
  - 파라미터 튜닝
    - num_leaves <= 2 ^ (max_depth) 
    - min_data_in_leaf : 100~9999, 언더피팅을 만들지도 모른다.
    - 속도향상
      - bagging_fraction / baggin_freq
      - feature_fraction
      - max_bin 작게
    - 정확도 향상
      - max_bin 크게
      - learning_rate 작게, num_iterations 크게
      - num_leaves 크게
      - dart
    - 과적합 조절
      - max_bin / num_leaves 작게
      - min_data_in_leaf / min_sum_hessian_in_leaf
      - bagging_fraction / bagging_freq
      - feature_fraction
      - lambda_l1 / lambda_l2 / min_gain_to_split 정규화

## 3. Regression 의 이해
  - 특정한 조건(가정)이 성립할 때, 의미를 갖는다.
    - Target Y 에 대해서 X 변수들은 독립적인 Linear 관계를 갖는다.
      - **X의 변화**에 Y는 반응해야한다.
      - 다른 X 변수와는 독립적이다.
      > 수학적으로 의미있는 최적화가 불가능, Plot으로 확인
    - **자기상관계수(Autocorrelation)**이 없어야한다.
      - 시간의 변화에 따른 상관관계
      > 추세를 확인, 깔때기 모양 Plot
    - **Correlation, multicollinearity**가 없어야한다.
        > VIF가 10에 가까우면 변수를 제거
    - 에러는 노멀분산을 가진다.
      - 깔때기 같은 분포, 예측의 분산이 커진다
      > log
